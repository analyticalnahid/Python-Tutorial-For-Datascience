## Exception Handling

Even though when your code has valid syntax, it may cause an error during execution.In Python, errors that occur during the execution are called exceptions.The causes of exceptions mainly come from the environment where the code executes.



```python
# An example of Single exception
# It will give error if divied by 0

try:  
    a = int(input("Enter a:"))    
    b = int(input("Enter b:"))    
    c = a/b  
    print(c)
except:  
    
    print("Can't divide with zero")  
```

```python
# An example of Multiple exception

try:      
    a=10/0      
except(ArithmeticError, IOError):      
    print("Arithmetic Exception")      
else:      
    print("Successfully Done")  
```

```python
# Finally
# The finally block will always be executed, no matter if the try block raises an error or not

try:
  x > 3
except:
  print("Something went wrong")
else:
  print("Nothing went wrong")
finally:
  print("The try...except block is finished")
```

```python
# Raise Exception

try:    
    age = int(input("Enter the age:"))    
    if(age<18):    
        raise ValueError   
    else:    
        print("the age is valid")    
except ValueError:    
    print("The age is not valid")   
```

### Python Assert

Python provides the assert statement to check if a given logical expression is true or false. Program execution proceeds only if the expression is true and raises the AssertionError when it is false. The following code shows the usage of the assert statement.

```python
try:
    num=int(input("Enter the even number"))
    assert num%2==0
    print("The number is even")
except AssertionError:
    
    print("Please enter even number")
```

## Inheritance

Inheritance allows us to define a class that inherits all the methods and properties from another class.
- Parent class is the class being inherited from, also called base class.
- Child class is the class that inherits from another class, also called derived class.



```python
# Parent Class

class Person:
    
  def __init__(self, fname, lname):
    self.firstname = fname
    self.lastname = lname

  def printname(self):
    print(self.firstname, self.lastname)

x = Person("John", "Doe")
x.printname()
```

```python
# Child Class 

class Student(Person):
  pass

x = Student("Mike", "Olsen")
x.printname()
```

## Encapsulation

The process of wrapping up variables and methods into a single entity is known as Encapsulation. It acts as a protective shield that puts restrictions on accessing variables and methods directly, and can prevent accidental or unauthorized modification of data.

- Protected Member (single underscore “_”)
- Private Member (double underscore “__”)

```python
# Protected members

class Base:
    def __init__(self):
 
        # Protected member
        self._a = 2
 
class Derived(Base):
    def __init__(self):
 
        # Calling constructor of Base class
        super.__init__(self)
        print("Calling protected member of base class: ",
              self._a)
 
        # Modify the protected variable:
        self._a = 3
        print("Calling modified protected member outside class: ",
              self._a)
 
 
obj1 = Derived()
obj2 = Base()

print("Accessing protected member of obj1: ", obj1._a)
print("Accessing protected member of obj2: ", obj2._a)
```

```python
# Private members

class Base:
    def __init__(self):
        self.a = "GeeksforGeeks"
        self.__c = "GeeksforGeeks"
 

class Derived(Base):
    def __init__(self):
 
        # Calling constructor of Base class
        super.__init__(self)
        print("Calling private member of base class: ")
        print(self.__c)
 
 
obj1 = Base()
print(obj1.a)
```

## Abstraction
An abstract class can be considered as a blueprint for other classes. It allows you to create a set of methods that must be created within any child classes built from the abstract class. A class which contains one or more abstract methods is called an abstract class

```python
from abc import ABC, abstractmethod
 
class Polygon(ABC):
 
    @abstractmethod
    def noofsides(self):
        pass

 
class Hexagon(Polygon):
 
    # overriding abstract method
    def noofsides(self):
        print("I have 6 sides")
 
class Quadrilateral(Polygon):
 
    # overriding abstract method
    def noofsides(self):
        print("I have 4 sides")

        

 
K = Hexagon()
K.noofsides()

R= Quadrilateral()
R.noofsides()
```

## Decorator

Decorator function is a function that receives another function as an argument. The behavior of the argument function is extended by the decorator without actually modifying it. The decorator function can be applied over a function using the @decorator syntax.

- property
- classmethod
- staticmethod

```python
# property decorator

class Portal:
  
    def __init__(self):
        self.__name =''
      
    @property
      
    # Getter method
    def name(self):
        return self.__name
      
    # Setter method
    @name.setter
    def name(self, val):
        self.__name = val
  
    # Deleter method
    @name.deleter
    def name(self):
        del self.__name
```

```python
p = Portal();
  
# Setting name
p.name = 'GeeksforGeeks'
  
# Prints name
print (p.name)
```

```python
# Deletes name
del p.name
```

```python
# As name is deleted above this 
# will throw an error
print (p.name)
```

## Memory Management

Memory management is the process of controlling and coordinating computer memory, assigning portions called blocks to various running programs to optimize overall system performance.

There are two parts of memory:
- stack memory
- heap memory



```python
# Multithreading

import threading
from time import time, sleep

start_time = time()

def work(id):
    print(f"Going to sleep...{id}")
    sleep(2) # Sleep for two Seconds
    print(f"Wake up...{id}")

# Simple Threads    
t1 = threading.Thread(target=work, args= [0])
t2 = threading.Thread(target=work,args= [1])

t1.start()
t2.start()

t1.join()
t2.join()


# Mores Threads in List
threads = [threading.Thread(target=work, args=[i]) for i in range(10)]

for thread in threads:
    thread.start()
    
for thread in threads:
    thread.join()

    
end_time = time()
print(f"Total time= {end_time-start_time} seconds")
```

```python
# Synchronizing Threads

balance = 200
lock = threading.Lock()

def deposit(amount, times, lock):
    global balance 
    
    for _ in range(times):
        lock.acquire()
        balance += amount
        lock.release()

def withdraw(amount, times, lock):
    global balance 
    
    for _ in range(times):
        lock.acquire()
        balance -= amount
        lock.release()
        
deposit_thread = threading.Thread(target= deposit, args= [1, 1000000, lock])
withdraw_thread = threading.Thread(target= withdraw , args = [1, 1000000, lock])

deposit_thread.start()
withdraw_thread.start()

deposit_thread.join()
withdraw_thread.join()

print(balance)
```

```python
# Threading Pool (An smart way to implement multithreading)
# Thread Pool using submit

from concurrent.futures import ThreadPoolExecutor, as_completed

start_time = time()

def work(id):
    print(f"Going to sleep...{id}")
    sleep(2)
    return (f"Wake up...{id}")

with ThreadPoolExecutor() as executor:
    task = [executor.submit(work, i) for i in range(10)]
    
    for _ in as_completed(task):
        print(_.result())

end_time = time()
print(f"Total time= {end_time-start_time} seconds")
```

```python
# Thread Pool using Map(which maintain sequence/order)

from concurrent.futures import ThreadPoolExecutor, as_completed

start_time = time()

def work(id):
    print(f"Going to sleep...{id}")
    sleep(2)
    return (f"Wake up...{id}")

ids = [0,1,2,3,4]
with ThreadPoolExecutor(max_workers=5) as executor:
    tasks = executor.map(work, ids)
    
    for task in tasks:
        print(task)

end_time = time()
print(f"Total time= {end_time-start_time} seconds")
```

```python
# Multiprocessing

from time import sleep, time
import multiprocessing
import os

start_time = time()

def work(id):
    print(f"Going to sleep...{id}")
    sleep(2)
    return (f"Wake up...{id}")

# Simple process
process1 = multiprocessing.Process(target= work, args=(0,))
process2 = multiprocessing.Process(target= work, args=(1,))

process1.start()
process2.start()

process1.join()
process2.join()

# Mores Process in List
processs = [multiprocessing.Process(target=work, args=(i,)) for i in range(10)]

for process in processs:
    process.start()
    
for process in processs:
    process.join()

end_time = time()
print(f"Total time= {end_time-start_time} seconds")
```

```python
# Process Pool (An smart way to implement multiprocessing)
# Process Pool using submit
# Note: It might not work on web, I suggest you to run on any ide eg: VScode, Pycharm

from concurrent.futures import ProcessPoolExecutor
from time import time, sleep

def work(id):
    print(f"Going to sleep...{id}")
    sleep(2)
    return (f"Wake up...{id}")

if __name__ == "__main__":
    print("Main Process Started")
    start_time = time()
    
    with ProcessPoolExecutor() as executor:
        tasks = [executor.submit(work, i) for i in range(10)]
        
        for task in tasks:
            print(task.result())
            
            
    end_time = time()       
    print(f"Total time= {end_time-start_time} seconds")
```
